```python
from cycler import cycler
from sklearn.model_selection import cross_val_predict
models = [
    ("Logistic Regression", LogisticRegression(C=31.845593832735297, penalty='l2')),
    ("Linear Discriminant Analysis", LinearDiscriminantAnalysis(solver='lsqr', shrinkage=0.0)),
    ("Bagging", BaggingClassifier(max_features=0.3462673894407039, max_samples=0.9582219015330582, n_estimators=200)),
    ("Random Forest", RandomForestClassifier(max_depth=25, min_samples_leaf=1, min_samples_split=20, n_estimators=155)),
    ("Gradient Boosting", GradientBoostingClassifier(learning_rate=0.05104619739803147, max_depth=10, min_samples_leaf=18, min_samples_split=17, n_estimators=103)),
    ("AdaBoost", AdaBoostClassifier(learning_rate=0.13218101548383096, n_estimators=200)),
    ("XGBoost", XGBClassifier(n_estimators=73, gamma=0.9277127751409531, max_depth=10, min_child_weight=20, subsample=0.9084822063853257, reg_alpha=0.009825784783514814, reg_lambda=4.262309566550185)),
    ("Extra Trees", ExtraTreesClassifier(max_depth=1135, min_samples_leaf=2, min_samples_split=2, n_estimators=1614))
]
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

plt.figure(figsize=(10,8))
plt.gca().set_prop_cycle(cycler('color', colors))
plt.plot([0,1],[0,1],'k--')

# Compute ROC curve and ROC area for each model
for name, model in models:
    y_scores = cross_val_predict(model, X_train, y_train, cv=5, method='predict_proba')
    fpr, tpr, _ = roc_curve(y_train, y_scores[:, 1])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=name)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic")
plt.legend(loc="lower right")
plt.grid(False)
plt.show()
```
