```python
from sklearn.ensemble import GradientBoostingClassifier, ExtraTreesClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
```

# Train

```python
# Gradient Boosting
gbdt = GradientBoostingClassifier()
gbdt.fit(x_train, y_train)
gbdt_accuracy_train = accuracy_score(y_train, gbdt.predict(x_train))
gbdt_auc_train = roc_auc_score(y_train, gbdt.predict_proba(x_train)[:, 1])
gbdt_accuracy_cv = cross_val_score(gbdt, x_train, y_train, cv=5)
gbdt_auc_cv = cross_val_score(gbdt, x_train, y_train, cv=5, scoring='roc_auc')
print(f'GBDT Accuracy (train): {gbdt_accuracy_train:.4f}')
print(f'GBDT AUC (train): {gbdt_auc_train:.4f}')
print(f'GBDT Accuracy (cv): {gbdt_accuracy_cv.mean():.4f}')
print(f'GBDT AUC (cv): {gbdt_auc_cv.mean():.4f}')
```

```python
# CatBoost
catboost = CatBoostClassifier()
catboost.fit(x_train, y_train)
catboost_accuracy_train = accuracy_score(y_train, catboost.predict(x_train))
catboost_auc_train = roc_auc_score(y_train, catboost.predict_proba(x_train)[:, 1])
catboost_accuracy_cv = cross_val_score(catboost, x_train, y_train, cv=5)
catboost_auc_cv = cross_val_score(catboost, x_train, y_train, cv=5, scoring='roc_auc')
print(f'CatBoost Accuracy (train): {catboost_accuracy_train:.4f}')
print(f'CatBoost AUC (train): {catboost_auc_train:.4f}')
print(f'CatBoost Accuracy (cv): {catboost_accuracy_cv.mean():.4f}')
print(f'CatBoost AUC (cv): {catboost_auc_cv.mean():.4f}')
```

```python
# Extra Trees
extra_trees = ExtraTreesClassifier()
extra_trees.fit(x_train, y_train)
extra_trees_accuracy_train = accuracy_score(y_train, extra_trees.predict(x_train))
extra_trees_auc_train = roc_auc_score(y_train, extra_trees.predict_proba(x_train)[:, 1])
extra_trees_accuracy_cv = cross_val_score(extra_trees, x_train, y_train, cv=5)
extra_trees_auc_cv = cross_val_score(extra_trees, x_train, y_train, cv=5, scoring='roc_auc')
print(f'Extra Trees Accuracy (train): {extra_trees_accuracy_train:.4f}')
print(f'Extra Trees AUC (train): {extra_trees_auc_train:.4f}')
print(f'Extra Trees Accuracy (cv): {extra_trees_accuracy_cv.mean():.4f}')
print(f'Extra Trees AUC (cv): {extra_trees_auc_cv.mean():.4f}')
```

```python
# Evaluating the performance of the GBDT model
gbdt_accuracy = round(accuracy_score(y_train, gbdt.predict(x_train)), 4)
gbdt_precision = round(precision_score(y_train, gbdt.predict(x_train)), 4)
gbdt_recall = round(recall_score(y_train, gbdt.predict(x_train)), 4)
gbdt_f1 = round(f1_score(y_train, gbdt.predict(x_train)), 4)

print("GBDT Model Performance:")
print("Accuracy: ", gbdt_accuracy)
print("Precision: ", gbdt_precision)
print("Recall: ", gbdt_recall)
print("F1 Score: ", gbdt_f1)
```

```python
# Evaluating the performance of the CatBoost model
catboost_accuracy = round(accuracy_score(y_train, catboost.predict(x_train)), 4)
catboost_precision = round(precision_score(y_train, catboost.predict(x_train)), 4)
catboost_recall = round(recall_score(y_train, catboost.predict(x_train)), 4)
catboost_f1 = round(f1_score(y_train, catboost.predict(x_train)), 4)

print("\nCatBoost Model Performance:")
print("Accuracy: ", catboost_accuracy)
print("Precision: ", catboost_precision)
print("Recall: ", catboost_recall)
print("F1 Score: ", catboost_f1)
```

```python
# Evaluating the performance of the Extra Trees model
extra_trees_accuracy = round(accuracy_score(y_train, extra_trees.predict(x_train)), 4)
extra_trees_precision = round(precision_score(y_train, extra_trees.predict(x_train)), 4)
extra_trees_recall = round(recall_score(y_train, extra_trees.predict(x_train)), 4)
extra_trees_f1 = round(f1_score(y_train, extra_trees.predict(x_train)), 4)

print("\nExtra Trees Model Performance:")
print("Accuracy: ", extra_trees_accuracy)
print("Precision: ", extra_trees_precision)
print("Recall: ", extra_trees_recall)
print("F1 Score: ", extra_trees_f1)
```

```python
y_pred_gbdt=gbdt.predict_proba(x_test)[:, 1]
y_pred_catboost=catboost.predict_proba(x_test)[:, 1]
y_pred_extra_trees=extra_trees.predict_proba(x_test)[:, 1]
```

```python
from skopt import BayesSearchCV, plots
from skopt.space import Real, Categorical, Integer

# Define the parameter space for GBDT
gbdt_param_space = {
    'learning_rate': Real(0.001, 0.01, prior='log-uniform'),
    'n_estimators': Integer(200, 800),
    'max_depth': Integer(3, 10),
}

# Create the GBDT model
gbdt = GradientBoostingClassifier()

# Perform Bayesian optimization for GBDT
opt_gbdt = BayesSearchCV(gbdt, gbdt_param_space, n_iter=50, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the GBDT model
opt_gbdt.fit(x_train, y_train)

# Get the best hyperparameters for GBDT
best_params_gbdt = opt_gbdt.best_params_
best_auc_gbdt = opt_gbdt.best_score_
print("Best hyperparameters for GBDT:", best_params_gbdt)
print("Best AUC for GBDT: %.4f" % best_auc_gbdt)
```

```python
# objective plot
plots.plot_objective(opt_gbdt.optimizer_results_[0])
```

```python
# evaluation plot
plots.plot_evaluations(opt_gbdt.optimizer_results_[0])
```

```python
# Define the parameter space for CatBoost
catboost_param_space = {
    'learning_rate': Real(0.001, 0.01, prior='log-uniform'),
    'n_estimators': Integer(200, 800),
    'depth': Integer(4, 8),
}

# Create the CatBoost model
catboost = CatBoostClassifier()

# Perform Bayesian optimization for CatBoost
opt_catboost = BayesSearchCV(catboost, catboost_param_space, n_iter=50, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the CatBoost model
opt_catboost.fit(x_train, y_train)

# Get the best hyperparameters for CatBoost
best_params_catboost = opt_catboost.best_params_
best_auc_catboost = opt_catboost.best_score_
print("Best hyperparameters for CatBoost:", best_params_catboost)
print("Best AUC for CatBoost: %.4f" % best_auc_catboost)
```

```python
# objective plot
plots.plot_objective(opt_catboost.optimizer_results_[0])
```

```python
# evaluation plot
plots.plot_evaluations(opt_catboost.optimizer_results_[0])
```

```python
# Define the parameter space for Extra Trees
extra_trees_param_space = {
    'n_estimators': Integer(200, 800),
    'max_depth': Integer(5, 15),
    'min_samples_split': Integer(2, 10),
}

# Create the Extra Trees model
extra_trees = ExtraTreesClassifier()

# Perform Bayesian optimization for Extra Trees
opt_extra_trees = BayesSearchCV(extra_trees, extra_trees_param_space, n_iter=50, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the Extra Trees model
opt_extra_trees.fit(x_train, y_train)

# Get the best hyperparameters for Extra Trees
best_params_extra_trees = opt_extra_trees.best_params_
best_auc_extra_trees = opt_extra_trees.best_score_
print("Best hyperparameters for Extra Trees:", best_params_extra_trees)
print("Best AUC for Extra Trees: %.4f" % best_auc_extra_trees)
```

```python
# objective plot
plots.plot_objective(opt_extra_trees.optimizer_results_[0])
```

```python
# evaluation plot
plots.plot_evaluations(opt_extra_trees.optimizer_results_[0])
```
